```python
import pandas as pd 
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### Fetching the Data 

```python
Adverstising_data=pd.read_csv('Advertising.csv')
```

```python
Adverstising_data.head()
```

```python
Adverstising_data.drop('Unnamed: 0',axis=1)
```

Here the KPI is sales so we are targetting to get the result for sales prediction 

```python
feature_cols = ['TV', 'radio', 'newspaper']
X=Adverstising_data[feature_cols]
y=Adverstising_data['sales']
```

# spliting the data in traing and testing

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
# default split is 75% for training and 25% for testing
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
print("Here is the Coff of Linear Regression",model.coef_)
```

```python
print("Here is the intercept of model ",model.intercept_)
```

for finding out the acuracy we generally take tesing data for predictions

```python
y_predt=model.predict(X_test)
```

```python
predict_Vs_actual_df=pd.DataFrame({'Predicted Value':y_predt,'Actual Value':y_test})
```

```python
predict_Vs_actual_df
```

# Now for Acuracy Point of view we will consider R2 MSE MAPE MAE MPE

```python
from sklearn import metrics 
print('MAE is ',metrics.mean_absolute_error(y_test,y_predt))
print('MSE is ',metrics.mean_squared_error(y_test,y_predt))
print('MAPE is ',np.mean(np.abs((y_test - y_predt) / y_test)) * 100,"%")
print('MPE is ',np.mean((y_test-y_predt)/y_test)*100)
print('r2 is ', metrics.r2_score(y_test, y_predt))
```
